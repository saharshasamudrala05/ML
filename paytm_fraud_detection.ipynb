Data Preprocessing :

Encoding of Type column

Dropping irrelevant columns like nameOrig, nameDest

Data Splitting

```python
new_type = pd.get_dummies(data['type'], drop_first= True)
new_data = pd.concat([data, new_type], axis=1)
new_data.head(5)
```

```python
X = new_data.drop(['isFraud', 'nameOrig', 'type', 'nameDest'], axis = 1)
y = new_data['isFraud']
```

```python
X.shape, y.shape
```

Split the data into Training and Testing data

```python
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(X,y, test_size=0.3, random_state=42)
```

Model Training

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score as ras
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
models = [
    LogisticRegression(),
    XGBClassifier(),
    RandomForestClassifier(n_estimators=7, criterion='entropy', random_state=7)
]

for i in range(len(models)):
    models[i].fit(X_train, y_train)
    print(f"{models[i]} : ")
    
    train_preds = models[i].predict_proba(X_train)[:, 1]
    print(f"Training accuracy: {ras(y_train, train_preds)}")
    
    y_preds = models[i].predict_proba(X_test)[:, 1]
    print(f"Validation accuracy: {ras(y_test, y_preds)}")
    print()
```

MODEL EVALUATION

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from matplotlib.colors import ListedColormap

y_pred = models[1].predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=models[1].classes_)

colors = ["#EE82EE", "#000000"]  # Violet and black
cmap = ListedColormap(colors)

disp.plot(cmap=cmap)
plt.show()



```
